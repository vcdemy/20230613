### Wide Form Example

### Long Form => Wide Form

```python
df.pivot('year','month','passengers')
```

```python
df.pivot?
```

```python
wide = df.pivot(index='year', columns='month', values='passengers')
```

```python
import plotly.express as px
```

```python
px.imshow(wide)
```

```python
px.imshow(wide, text_auto=True)
```

```python
wide = wide[['January','February','March','April','May','June',
             'July','August','September','October','November','December']]
```

```python
px.imshow(wide, text_auto=True)
```

### Wide Form => Long Form

```python
wide
```

```python
wide.melt?
```

```python
wide.reset_index()
```

```python
wide.melt?
```

```python
wide.reset_index().melt(id_vars='year',var_name='month',value_name='passengers')
```

### .groupby()

```python
import numpy as np
```

```python
np.random.seed(987)
ids = np.random.choice(['A','B','C'], 20)
values = np.random.randint(1, 10, 20)
```

```python
df = pd.DataFrame({'id':ids, 'value':values})
```

```python
df
```

```python
grps = df.groupby('id')
```

```python
grps.sum()
```

```python
grps.get_group('A')
```

```python
grps.get_group('B')
```

```python
grps.mean()
```

```python
grps.count()
```

```python
grps.agg(['sum','mean','count'])
```

### Filtering / Subsetting / Boolean Indexing

```python
df
```

```python
df['國文'] >= 60
```

### 列出國文及格的同學

```python
df[df['國文'] >= 60]
```

```python
df[df['國文'] >= 60].index
```

```python
(df['國文'] >= 60) & (df['英文'] >= 60)
```

### 列出國文跟英文都及格的同學

```python
df[(df['國文'] >= 60) & (df['英文'] >= 60)]
```

### 列出國文或英文及格的同學

```python
df[(df['國文'] >= 60) | (df['英文'] >= 60)]
```
